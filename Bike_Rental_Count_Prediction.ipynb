```python
rent_df1 = rent_df1.drop(columns = ['Date','Year','holiday','weekday','workingday','Weather','casual','registered'])
rent_df1.head()
#Deleted Unwanted Columns
```

```python
#Preparing Dataset for Training
from sklearn.model_selection import train_test_split 
X = rent_df1.drop(['Humidity','windspeed','Total_Count'], axis = "columns")
Y0 = rent_df1['Total_Count']
Y2 = rent_df1['Humidity']
Y3 = rent_df1['windspeed']
X.head()
```

```python
#Dividing the Data into Training and Test using sklearn's Library
X_train,X_test,Y0_train,Y0_test,Y2_train,Y2_test,Y3_train,Y3_test=train_test_split(X,Y0,Y2,Y3,test_size=.25,random_state=0)
```

#**DECISION TREE**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics

#Lets Train the Decision Tree Algorithm
DTModel_Total_Count=DecisionTreeRegressor(max_depth=2).fit(X_train,Y0_train)
DTModel_Humidity=DecisionTreeRegressor(max_depth=2).fit(X_train,Y2_train)
DTModel_windspeed=DecisionTreeRegressor(max_depth=2).fit(X_train,Y3_train)

#Evaluating the Algorithm
y0_pred = DTModel_Total_Count.predict(X_test)
y2_pred = DTModel_Humidity.predict(X_test)
y3_pred = DTModel_windspeed.predict(X_test)

print("Accuracy_Total_Count = ",DTModel_Total_Count.score(X_train,Y0_train))
print("Accuracy_Humidity    = ",DTModel_Humidity.score(X_train,Y2_train))
print("Accuracy_windspeed   = ",DTModel_windspeed.score(X_train,Y3_train))
print('\n')
print('Mean Absolute Error Total_Count    : ', metrics.mean_absolute_error(Y0_test, y0_pred))
print('Mean Squared Error Total_Count     : ', metrics.mean_squared_error(Y0_test, y0_pred))
print('Root Mean Squared Error Total_Count: ', np.sqrt(metrics.mean_squared_error(Y0_test, y0_pred)))
print('\n')
print('Mean Absolute Error Humidity    : ', metrics.mean_absolute_error(Y2_test, y2_pred))
print('Mean Squared Error Humidity     : ', metrics.mean_squared_error(Y2_test, y2_pred))
print('Root Mean Squared Error Humidity: ', np.sqrt(metrics.mean_squared_error(Y2_test, y2_pred)))
print('\n')
print('Mean Absolute Error windspeed    : ', metrics.mean_absolute_error(Y3_test, y3_pred))
print('Mean Squared Error windspeed     : ', metrics.mean_squared_error(Y3_test, y3_pred))
print('Root Mean Squared Error windspeed: ', np.sqrt(metrics.mean_squared_error(Y3_test, y3_pred)))
```

#**RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

#Lets Train the Random Forest Algorithm
RFModel_Total_Count=RandomForestRegressor(n_estimators=200).fit(X_train,Y0_train)
RFModel_Humidity=RandomForestRegressor(n_estimators=200).fit(X_train,Y2_train)
RFModel_windspeed=RandomForestRegressor(n_estimators=200).fit(X_train,Y3_train)

#Evaluating the Algorithm
y0_pred = RFModel_Total_Count.predict(X_test)
y2_pred = RFModel_Humidity.predict(X_test)
y3_pred = RFModel_windspeed.predict(X_test)

print("Accuracy_Total_Count = ",RFModel_Total_Count.score(X_train,Y0_train))
print("Accuracy_Humidity    = ",RFModel_Humidity.score(X_train,Y2_train))
print("Accuracy_windspeed   = ",RFModel_windspeed.score(X_train,Y3_train))
print('\n')
print('Mean Absolute Error Total_Count    : ', metrics.mean_absolute_error(Y0_test, y0_pred))
print('Mean Squared Error Total_Count     : ', metrics.mean_squared_error(Y0_test, y0_pred))
print('Root Mean Squared Error Total_Count: ', np.sqrt(metrics.mean_squared_error(Y0_test, y0_pred)))
print('\n')
print('Mean Absolute Error Humidity    : ', metrics.mean_absolute_error(Y2_test, y2_pred))
print('Mean Squared Error Humidity     : ', metrics.mean_squared_error(Y2_test, y2_pred))
print('Root Mean Squared Error Humidity: ', np.sqrt(metrics.mean_squared_error(Y2_test, y2_pred)))
print('\n')
print('Mean Absolute Error windspeed    : ', metrics.mean_absolute_error(Y3_test, y3_pred))
print('Mean Squared Error windspeed     : ', metrics.mean_squared_error(Y3_test, y3_pred))
print('Root Mean Squared Error windspeed: ', np.sqrt(metrics.mean_squared_error(Y3_test, y3_pred)))
```

#**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics

#Lets Train the Linear Regression Algorithm
LRModel_Total_Count= LinearRegression().fit(X_train,Y0_train)
LRModel_Humidity= LinearRegression().fit(X_train,Y2_train)
LRModel_windspeed= LinearRegression().fit(X_train,Y3_train)

#Evaluating the Algorithm
y0_pred = LRModel_Total_Count.predict(X_test)
y2_pred = LRModel_Humidity.predict(X_test)
y3_pred = LRModel_windspeed.predict(X_test)

print("Accuracy_Total_Count = ",LRModel_Total_Count.score(X_train,Y0_train))
print("Accuracy_Humidity    = ",LRModel_Humidity.score(X_train,Y2_train))
print("Accuracy_windspeed   = ",LRModel_windspeed.score(X_train,Y3_train))
print('\n')
print('Mean Absolute Error Total_Count    : ', metrics.mean_absolute_error(Y0_test, y0_pred))
print('Mean Squared Error Total_Count     : ', metrics.mean_squared_error(Y0_test, y0_pred))
print('Root Mean Squared Error Total_Count: ', np.sqrt(metrics.mean_squared_error(Y0_test, y0_pred)))
print('\n')
print('Mean Absolute Error Humidity    : ', metrics.mean_absolute_error(Y2_test, y2_pred))
print('Mean Squared Error Humidity     : ', metrics.mean_squared_error(Y2_test, y2_pred))
print('Root Mean Squared Error Humidity: ', np.sqrt(metrics.mean_squared_error(Y2_test, y2_pred)))
print('\n')
print('Mean Absolute Error windspeed    : ', metrics.mean_absolute_error(Y3_test, y3_pred))
print('Mean Squared Error windspeed     : ', metrics.mean_squared_error(Y3_test, y3_pred))
print('Root Mean Squared Error windspeed: ', np.sqrt(metrics.mean_squared_error(Y3_test, y3_pred)))
```

**Final model for predicting the bike rental count on daily basis:**
When we compare the root mean squared error and mean absolute error of all 3 models, the random forest model has more accuracy. So, finally random forest model is best for predicting the bike rental count on daily basis.
